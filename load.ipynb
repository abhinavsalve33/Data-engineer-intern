```python
import pandas as pd
import mysql.connector

# Load cleaned data
path = "C:\\Users\\hp\\Downloads\\assingment\\hackernews_data_clean.csv"
df = pd.read_csv(path)

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="HackerDB1"
)

cursor = conn.cursor()

# Insert query
query = """
INSERT INTO hackernews_data (title, link)
VALUES (%s, %s)
"""

# Insert rows
for index, row in df.iterrows():
    cursor.execute(query, (
        row["title"],
        row["link"]
    ))

# Commit changes
conn.commit()

print(cursor.rowcount, "records inserted.")

# Close connection
cursor.close()
conn.close()
```
