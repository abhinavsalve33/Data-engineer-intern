```python
# !pip install fastapi uvicorn
# !pip install psycopg2-binary
from fastapi import FastAPI
import psycopg2

app = FastAPI()

@app.get("/jobs")
def get_jobs():

    conn = psycopg2.connect(
        host="localhost",
        database="HackerDB1",
        user="root",
        password="root"
    )

    cur = conn.cursor()

    cur.execute("SELECT * FROM hackernews_data LIMIT 50")

    rows = cur.fetchall()

    cur.close()
    conn.close()

    return rows
```
