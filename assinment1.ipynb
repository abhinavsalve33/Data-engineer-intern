```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

news_data = []

url = "https://news.ycombinator.com/"

response = requests.get(url)

soup = BeautifulSoup(response.text,"html.parser")

titles = soup.find_all("span",class_="titleline")

for title in titles:

    link = title.find("a")

    news_data.append({
        "title": link.text if link else None,
        "link": link["href"] if link else None
    })

df = pd.DataFrame(news_data)

df.to_csv("hackernews_data.csv",index=False)

print("News data saved")
```

```python
import pandas as pd

df = pd.read_csv("hackernews_data.csv")

df.dropna(inplace=True)


df.drop_duplicates(inplace=True)


df["title"] = df["title"].str.strip()


df = df[df["link"].str.startswith("http")]


df.reset_index(drop=True, inplace=True)
df.to_csv("hackernews_data_clean.csv", index=False)

print(" data saved")
```
